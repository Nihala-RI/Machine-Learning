```python
x=df.drop('class',axis=1)
y=df['class']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
ann=tf.keras.models.Sequential()
ann.add(tf.keras.layers.Dense(units=20,activation='relu'))
ann.add(tf.keras.layers.Dense(units=20,activation='relu'))
ann.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
```

```python
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
ann.fit(x_train,y_train,batch_size=8,epochs=10)
```

```python
ann.evaluate(x_test,y_test)
```

```python
y_pred=ann.predict(sc.fit_transform(x_test))
```

```python
y_pred
```

```python
y_prediction=(y_pred>0.5).ravel()
```

```python
pd.DataFrame({'Actual Class':y_test,'Predicted Class':y_prediction})
```

```python
y_prediction=[]
for i in y_pred:
  if i>0.5:
    y_prediction.append(1)
  else:
    y_prediction.append(0)
print(y_prediction)
```

```python
y_predictions=[]
for i in y_pred:
  if i>0.5:
    y_predictions.append('Poisonous')
  else:
    y_predictions.append('Edible')
print(y_predictions)
```

```python
pd.DataFrame({'Actual Class':y_test,'Predicted Class':y_predictions})
```

```python
a=pd.DataFrame(y_predictions)
class_names=a[0].unique()
class_names
```

```python
from sklearn.metrics import confusion_matrix,classification_report,roc_auc_score,roc_curve
cm=confusion_matrix(y_test,y_prediction)
plt.figure(figsize=(17,7))
sns.heatmap(cm,annot=True,fmt='d',xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Actual_values')
plt.ylabel('Predicted_values')
```

```python
print(classification_report(y_test,y_prediction,target_names=class_names))
```

```python
# generate a no skill prediction (majority class)
ns_probs = [0 for _ in range(len(y_test))]
def plot_roc_curve(y_test,y_pred):
  ns_fpr, ns_tpr, _ = roc_curve(y_test, ns_probs)
  fpr,tpr,thresholds=roc_curve(y_test,y_prediction)
  plt.plot(ns_fpr,ns_tpr,linestyle='--', label='No skill')
  plt.plot(fpr,tpr,marker='.', label='ANN')
  plt.xlabel('False Positive Rate')
  plt.ylabel('True Positive Rate')
  plt.legend()
  plt.show()
  print(f"ANN:",{roc_auc_score(y_test,y_prediction)})
  print(f"No Skill:",{roc_auc_score(y_test,ns_probs)})
```

```python
plot_roc_curve(y_test,y_prediction)
```
